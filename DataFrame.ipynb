```python
import pandas as pd
import random

# Generate data
names = [f"Student_{i}" for i in range(1, 51)]
roll_numbers = [f"RN{i:03d}" for i in range(1, 51)]
ages = [random.randint(18, 25) for _ in range(50)]
subjects = ["Math", "Science", "English", "History", "Geography"]

# Create subject-marks column with format like "Math-85"
subject_marks = [f"{random.choice(subjects)}-{random.randint(50, 100)}" for _ in range(50)]

# Create DataFrame
df = pd.DataFrame({
    "name": names,
    "roll number": roll_numbers,
    "age": ages,
    "subject-marks": subject_marks
})

# Display DataFrame
df


```

```python
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}")
        print(f"Roll: {self.roll}")
        print(f"Marks: {self.marks}")

s1 = Student("Riya", 101, 89)
s1.display()
```

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):  # Overriding the parent method
        print("Dog barks")

a = Animal()
a.speak()  # Output: Animal makes a sound

d = Dog()
d.speak()  # Output: Dog barks  ← (Overridden)

```
